```python
import pandas as pd
import seaborn as sns

raw_data: pd.DataFrame = pd.read_csv('data/fake_job_postings.csv')
raw_data
```

```python
# Create new dataframe with concatenated text columns and fraudulent label
data = pd.DataFrame({
  'job_description': raw_data[['company_profile', 'description', 'requirements', 'benefits']].fillna('').agg(' '.join, axis=1),
  'fraudulent': raw_data['fraudulent']
})
data
```

## Data Exploration

Plot a histogram of the lengths of the `job_description` for fraudulent and non-fraudulent job postings

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(17, 5), dpi=100)
length = data[data["fraudulent"] == 1]["job_description"].str.len()
ax1.hist(length, bins=20, color="orangered")

ax1.set_title("Fake Post")
length = data[data["fraudulent"] == 0]["job_description"].str.len()
ax2.hist(length, bins=20)
ax2.set_title("Real Post")
fig.suptitle("Characters in description")
plt.show()
```

```python
# Average description length by fraudulent status
data['description_length'] = data['job_description'].str.len()
data.groupby('fraudulent')['description_length'].mean()
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    data['job_description'], data['fraudulent'], test_size=0.2, random_state=42
)

# Oversampling the minority class in the training set
ros = RandomOverSampler(random_state=42)
X_train_resampled, y_train_resampled = ros.fit_resample(X_train.to_frame(), y_train)
X_train = X_train_resampled['job_description']
y_train = y_train_resampled
# Checking new class distribution
print(y_train_resampled.value_counts())
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define a function to plot side-by-side histograms
def plot_histogram_side_by_side(column, title, max_labels=10):
    # Count occurrences of each category
    counts = raw_data[column].value_counts()

    # Ensure `top_categories` is always defined
    top_categories = counts.index[:max_labels] if len(counts) > max_labels else counts.index

    raw_data_filtered = raw_data[raw_data[column].isin(top_categories)]

    # Split fraudulent and non-fraudulent data
    non_fraud_data = raw_data_filtered[raw_data_filtered["fraudulent"] == 0][column]
    fraud_data = raw_data_filtered[raw_data_filtered["fraudulent"] == 1][column]

    # Create side-by-side plots
    fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

    # Non-Fraudulent Plot
    axes[0].hist(non_fraud_data, bins=len(top_categories), color="blue", alpha=0.7)
    axes[0].set_title(f"Non-Fraudulent {title}")
    axes[0].set_xlabel(title)
    axes[0].set_ylabel("Frequency")

    # Fraudulent Plot
    axes[1].hist(fraud_data, bins=len(top_categories), color="red", alpha=0.7)
    axes[1].set_title(f"Fraudulent {title}")
    axes[1].set_xlabel(title)

    # Rotate x-axis labels properly (slanted for both)
    for ax in axes:
        ax.set_xticks(range(len(top_categories)))
        ax.set_xticklabels(top_categories, rotation=45, ha="right")

    plt.tight_layout()
    plt.show()

# Plot histograms for 'location' and 'required_experience' side by side
plot_histogram_side_by_side("location", "Location", max_labels=10)
plot_histogram_side_by_side("required_experience", "Required Experience", max_labels=8)


# Special handling for salary range (numerical or categorical)
def plot_salary_histogram_side_by_side():
    # Check if salary range is categorical or numerical
    if raw_data["salary_range"].dtype == "object":
        salary_counts = raw_data["salary_range"].value_counts()
        top_salaries = salary_counts.index[:10] if len(salary_counts) > 10 else salary_counts.index
        raw_data_filtered = raw_data[raw_data["salary_range"].isin(top_salaries)]

        non_fraud_data = raw_data_filtered[raw_data_filtered["fraudulent"] == 0]["salary_range"]
        fraud_data = raw_data_filtered[raw_data_filtered["fraudulent"] == 1]["salary_range"]

        fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

        # Non-Fraudulent Salary Plot
        axes[0].hist(non_fraud_data, bins=len(top_salaries), color="blue", alpha=0.7)
        axes[0].set_title("Non-Fraudulent Salary Range")
        axes[0].set_xlabel("Salary Range")

        # Fraudulent Salary Plot
        axes[1].hist(fraud_data, bins=len(top_salaries), color="red", alpha=0.7)
        axes[1].set_title("Fraudulent Salary Range")
        axes[1].set_xlabel("Salary Range")

        # Rotate x-axis labels properly (slanted for both)
        for ax in axes:
            ax.set_xticks(range(len(top_salaries)))
            ax.set_xticklabels(top_salaries, rotation=45, ha="right")

    else:  # If numerical, use histogram bins
        bins = np.linspace(raw_data["salary_range"].min(), raw_data["salary_range"].max(), 10)  # 10 bins
        fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

        # Non-Fraudulent Salary Plot
        axes[0].hist(raw_data[raw_data["fraudulent"] == 0]["salary_range"], bins=bins, color="blue", alpha=0.7)
        axes[0].set_title("Non-Fraudulent Salary Range")
        axes[0].set_xlabel("Salary Range")

        # Fraudulent Salary Plot
        axes[1].hist(raw_data[raw_data["fraudulent"] == 1]["salary_range"], bins=bins, color="red", alpha=0.7)
        axes[1].set_title("Fraudulent Salary Range")
        axes[1].set_xlabel("Salary Range")

        # Reduce x-axis label clutter
        for ax in axes:
            ax.set_xticks(bins.round(2)[::2])  # Show only every second label
            ax.tick_params(axis="x", rotation=45)  # Rotate labels to slant them

    plt.tight_layout()
    plt.show()

# Call function to plot salary range
plot_salary_histogram_side_by_side()
```

Use `TfidfVectorizer` to convert text data into vectors

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english', max_features=60000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

print(X_train_vec.shape)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
def evaluate(y_test, y_pred):
  print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
  print(f'Confusion matrix:\n {confusion_matrix(y_test, y_pred)}')
  print(f"{classification_report(y_test, y_pred)}")
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train_vec, y_train)
evaluate(y_test, model.predict(X_test_vec))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train_vec, y_train)
evaluate(y_test, model.predict(X_test_vec))
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_vec, y_train)
evaluate(y_test, model.predict(X_test_vec))
```

```python
from sklearn.linear_model import SGDClassifier
model = SGDClassifier()
model.fit(X_train_vec, y_train)

evaluate(y_test, model.predict(X_test_vec))
```
